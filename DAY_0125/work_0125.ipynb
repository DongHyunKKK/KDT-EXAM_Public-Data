```python
subway_line = ['1호선', '2호선', '3호선', '4호선', '5호선', '6호선', '7호선']
timeDF = timeDF[timeDF['호선명'].isin(subway_line)]
```

```python
timeDF.head()
```

```python
timeDF.info()
```

```python
timeDF.dtypes
```

```python
timeDF[timeDF.columns[11]] = timeDF[timeDF.columns[11]].astype('int64')
timeDF[timeDF.columns[13]] = timeDF[timeDF.columns[13]].astype('int64')
```

```python
timeDF.dtypes
```

```python
timeDF
```

```python
# 데이터프레임 객체를 얻기 위해 max()함수 외에 min() 함수 써도 된다.
maxDF = timeDF.groupby(['호선명', '지하철역'])[[timeDF.columns[11], timeDF.columns[13]]].max() 
maxDF
```

```python
maxDF.index
```

```python
lineDF = maxDF.xs('1호선', level = 0)
lineDF
```

```python
lineDF = lineDF[lineDF.columns[0]] + lineDF[lineDF.columns[1]]
lineDF
```

```python
lineDF.max()
```

```python
lineDF.idxmax()
```

```python
max_num = []
station_list = []
xticks_list = []
for line in subway_line:
    lineDF = maxDF.xs(line, level = 0)
    lineDF = lineDF[lineDF.columns[0]] + lineDF[lineDF.columns[1]]
    station = lineDF.idxmax()
    station_list.append(station)
    max_num.append(lineDF.max())
    xticks_list.append('{} {}'.format(line, station))
    print(f'출근 시간대 {line} 최대 하차역 : {station}역, 하차인원 : {lineDF.max():,}명')

plt.figure(figsize = (10, 5))
plt.bar(xticks_list, max_num)
plt.xticks(rotation = 80)
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역')
plt.show()
```

- 문제 2    

DataFrame 사용

```python
timeDF = pd.read_csv(file)
timeDF.head()
```

```python
timeDF.drop(0, inplace = True)
```

```python
timeDF[timeDF.columns[35]] = timeDF[timeDF.columns[35]].astype('int64')
timeDF[timeDF.columns[37]] = timeDF[timeDF.columns[37]].astype('int64')
```

```python
timeDF1 = timeDF[['지하철역', timeDF.columns[35], timeDF.columns[37]]]
timeDF1
```

```python
timeDF1['지하철역'].value_counts()
```

```python
timeDF1['sum'] = timeDF1[timeDF1.columns[1]] + timeDF1[timeDF1.columns[2]]
```

```python
df = timeDF1.groupby('지하철역')['sum'].sum()
df
```

```python
# groupby 메서드를 사용하면 중복인덱스는 사라진다.
stations= []
for name in list(df.index):
    print(f'{name} : 인덱스에서 {list(df.index).count(name)}개 있다.')
    if list(df.index).count(name) > 1:
        stations.append(name)
stations
```

```python
df = df.sort_values(ascending = False)
```

```python
df
```

```python
station_list = []
passengers = []
for i in range(5):
   station_list.append(df.index[i])
   passengers.append(df.values[i])
   print(f'{df.index[i]} : {df.values[i]:,}')

plt.figure(figsize = (10, 5))
plt.bar(station_list, passengers)
plt.title('서울 지하철 퇴근 시간대 하차 인원 비교')
plt.show()
```
