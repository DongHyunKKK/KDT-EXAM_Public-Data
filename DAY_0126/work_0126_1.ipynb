CSV 사용

```python
file = '../DATA/gender.csv'
```

```python
# subplot 사용
f = open(file, encoding = 'utf-8-sig')
data = csv.reader(f)
population = []
city = '대구광역시'
male_count = 0
female_count = 0
label_list = ['남성', '여성']

plt.figure(figsize = (20, 20))
plt.suptitle('대구광역시 구별 남녀 인구 비율', size = 20)
i = 1
for row in data:
    if city in row[0] and i < 10:
        male_count = int(row[104].replace(',', ''))  # 자리수 문자열 제거 및 숫자로 변환
        female_count = int(row[207].replace(',', ''))
        population_list = [male_count, female_count]
        plt.subplot(4, 3, i)
        plt.pie(population_list, labels = label_list, autopct = '%.1f%%', startangle = 90)
        plt.title(row[0], size = 15)
        i += 1

plt.show()
```

```python
# subplots 사용
f = open(file, encoding = 'utf-8-sig')
data = csv.reader(f)
population = []
city = '대구광역시'
male_count = 0
female_count = 0
label_list = ['남성', '여성']

fig, ax = plt.subplots(3, 3, figsize = (15, 15))
plt.suptitle('대구광역시 구별 남녀 인구 비율', size = 20)
i = 1
for row in data:
    if city in row[0] and i < 10:
        male_count = int(row[104].replace(',', ''))  # 자리수 문자열 제거 및 숫자로 변환
        female_count = int(row[207].replace(',', ''))
        population_list = [male_count, female_count]
        ax[(i-1) // 3, (i-1) % 3].pie(population_list, labels = label_list, autopct = '%.1f%%', startangle = 90)
        ax[(i-1) // 3, (i-1) % 3].set_title(row[0], size = 15)
        i += 1
        
plt.show()
```
